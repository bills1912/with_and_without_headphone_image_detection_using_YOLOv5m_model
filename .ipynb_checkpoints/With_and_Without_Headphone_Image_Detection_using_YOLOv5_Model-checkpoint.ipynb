## Import Package Needed

```python
import os
import cv2
import time
import uuid   # Unique identifier
import torch
import numpy as np
from matplotlib import pyplot as plt
```

## Collect Train and Test Data

```python
IMAGES_PATH_TRAIN = os.path.join('headphone', 'images') #/data/images
labels_train = ['without_headphone', 'with_headphone']
number_imgs_train = 30
```

```python
CAPTURE = cv2.VideoCapture(0)
for label in labels_train:
    print('Collecting images for {}'.format(label))
    time.sleep(5)

    for img_num in range(number_imgs_train):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        ret, frame = CAPTURE.read()
        imgname = os.path.join(IMAGES_PATH_TRAIN, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
CAPTURE.release()
cv2.destroyAllWindows()
```

```python
IMAGES_PATH_TEST = os.path.join('headphone_test', 'images') #/data/images
labels_test = ['without_headphone', 'with_headphone']
number_imgs_test = 10
```

```python
CAPTURE = cv2.VideoCapture(0)
for label in labels_test:
    print('Collecting images for {}'.format(label))
    time.sleep(5)

    for img_num in range(number_imgs_test):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        ret, frame = CAPTURE.read()
        imgname = os.path.join(IMAGES_PATH_TEST, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
CAPTURE.release()
cv2.destroyAllWindows()
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp21/weights/last.pt', force_reload=True)
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
